```python
# Now get the table for shooting stats

sample_shooting_df = pd.read_html(io.StringIO(shooting_page.text), match="Shooting ")
sample_shooting_df = sample_shooting_df[0]
sample_shooting_df.head()
```

```python
# We already have many of the columns in our stats dataframe. So we need only specific columns in this dataframe.
columns = ["Date", "Comp", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]
sample_shooting_df = sample_shooting_df[columns]
sample_shooting_df.head()

# Next we need to merge this dataframe with the stats dataframe. 
```

```python
def data_scraper(team_urls):
    all_dfs = []
    for url in team_urls:
        team_name = url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        
        data = requests.get(url)
        dataframe_score = pd.read_html(io.StringIO(data.text), match="Scores & Fixtures ")[0]

        soup = BeautifulSoup(data.text)
        a_tags = soup.find_all("a")
        href_links = [tag.get("href") for tag in a_tags]
        shooting_page_link = [link for link in href_links if link and "all_comps/shooting/" in link]
        shooting_page_link = f"https://fbref.com{shooting_page_link[0]}"

        time.sleep(2)
        shooting_page_data = requests.get(shooting_page_link)

        dataframe_shoot = pd.read_html(io.StringIO(shooting_page_data.text), match="Shooting ")[0]
        dataframe_shoot.columns = dataframe_shoot.columns.droplevel(0)

        columns = ["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]
        dataframe_shoot = dataframe_shoot[columns]

        try:
            df = dataframe_score.merge(dataframe_shoot, on=["Date"])
        except ValueError:
            continue

        df = df[df["Comp"] == "La Liga"]
        df["Team"] = team_name
        all_dfs.append(df)
        
        time.sleep(2)

    
    return all_dfs
```

```python
df_list = data_scraper(team_urls)
```

```python
df = pd.concat(df_list)
```
